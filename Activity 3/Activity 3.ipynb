```python
#Import data file
import pandas as pd
data = pd.read_csv("mutations.csv")
data.rename(columns={"Unnamed: 0": "Individual"}, inplace=True)
#data
```

```python
#This cell is all of Activity 2 from last week (with my corrections to the %C and %NC columns)


#Calculate totals for each row and column and load it into a dataframe
#Summing up the rows, keeping the individual labels in column 1
num_mut_per_ind = data.sum(1)
df1 = num_mut_per_ind.to_frame()
df1.insert(0, "Individual", data['Individual'])
df1.rename(columns={0 : "Total Mutations"}, inplace=True)
#Summing up the columns, keeping the mutations in column 1
num_ind_per_mut = data.sum()
df2 = num_ind_per_mut.to_frame()
df2.reset_index(inplace=True)
df2.index -= 1
df2.rename(columns={"index": "Mutation", 0 : "Total Individuals"}, inplace=True)
df2 = df2[1:1411]
# df1
# df2

#Plot number of mutations per individual
import matplotlib.pyplot as plt
plt.scatter(df1.index, df1.loc[:,"Total Mutations"])
plt.title("Number of Mutations Per Individual")
plt.xlabel("Individual")
plt.ylabel("Mutations")
plt.show()

#Plot number of individuals per mutation
plt.scatter(df2.index, df2.loc[:,"Total Individuals"])
plt.title("Number of Individuals Per Mutation")
plt.xlabel("Mutation")
plt.ylabel("Individuals")
plt.show()

#Construct a feature table with additional analysis on the data
feature_table_df = df2.copy()
#First column T = total individuals per mutation is already calculated
feature_table_df.rename(columns={"Mutation" : "Genetic Mutation (Feature)", "Total Individuals" : 'T'}, inplace=True)
# C = number of individuals with cancer with that mutation
feature_table_df['C'] = 0
# NC = number of non-cancer individuals with that mutation
feature_table_df['NC'] = 0
# %C = % of total individuals with that mutation with cancer
feature_table_df['%C'] = 0
# %NC = % of total individuals with that mutation without cancer
feature_table_df['%NC'] = 0
# %C - %NC = Difference between %C and %NC
feature_table_df['%C - %NC'] = 0
# %C / %NC = Ratio of %C to %NC
feature_table_df['%C / %NC'] = 0
#feature_table_df

#Count up how many individuals with cancer have each mutation
#Axis 0 = vertical axis
for x in range(len(data.axes[0])):
    #Axis 1 = horizontal axis
    for y in range(len(data.axes[1])):
        #Check for mutation and if the individual has cancer
        if data.iloc[x,y] == 1 and data.iloc[x,0][:1] == 'C':
            feature_table_df.loc[y-1,'C'] += 1
            
#Get total number of cancer and non-cancer individuals
total_c = 0
for x in range(len(data.axes[0])):
    #Check if individual is a C
    if data.iloc[x,0][:1] == 'C':
        total_c += 1
        
total_nc = len(data.axes[0]) - total_c

# print(total_c)
# print(total_nc)

#From that information, the rest of the table can be calculated
for y in range(len(feature_table_df.axes[0])):
    feature_table_df.loc[y,'NC'] = feature_table_df.loc[y,'T'] - feature_table_df.loc[y,'C']
    feature_table_df.loc[y,'%C'] = feature_table_df.loc[y,'C'] / total_c * 100
    feature_table_df.loc[y,'%NC'] = feature_table_df.loc[y,'NC'] / total_nc * 100
    feature_table_df.loc[y,'%C - %NC'] = feature_table_df.loc[y,'%C'] - feature_table_df.loc[y,'%NC']
    if feature_table_df.loc[y,'%NC'] != 0:
        feature_table_df.loc[y,'%C / %NC'] = feature_table_df.loc[y,'%C'] / feature_table_df.loc[y,'%NC']
    else:
        #Use some small number to avoid dividing by 0
        feature_table_df.loc[y,'%C / %NC'] = feature_table_df.loc[y,'%C'] / 0.000001

#Write out the data to a csv
# feature_table_df.to_csv("feature_table.csv", index=True)

#Format the data to display more clearly here
pd.options.display.float_format = '{:.2f}'.format
feature_table_df = feature_table_df.astype({"T":'int', "C":'int', "NC":'int'}) 
print("feature_table_df : ")
display(feature_table_df)

#Construct a summary feature table, showing the top 10 features (mutations) of each column
#First column, sorted by T
col1_df = pd.DataFrame(feature_table_df.sort_values('T', ascending=False).head(10))
print("col1_df : ")
display(col1_df)

#Second column, sorted by C
col2_df = pd.DataFrame(feature_table_df.sort_values('C', ascending=False).head(10))
print("col2_df : ")
display(col2_df)

#Third column, sorted by NC
col3_df = pd.DataFrame(feature_table_df.sort_values('NC', ascending=False).head(10))
print("col3_df : ")
display(col3_df)

#Fourth column, sorted by %C
col4_df = pd.DataFrame(feature_table_df.sort_values('%C', ascending=False).head(10))
print("col4_df : ")
display(col4_df)

#Fifth column, sorted by %NC
col5_df = pd.DataFrame(feature_table_df.sort_values('%NC', ascending=False).head(10))
print("col5_df : ")
display(col5_df)

#Sixth columnm, sorted by %C - %NC
col6_df = pd.DataFrame(feature_table_df.sort_values('%C - %NC', ascending=False).head(10))
print("col6_df : ")
display(col6_df)

#Seventh column, sorted by %C / %NC
col7_df = pd.DataFrame(feature_table_df.sort_values('%C / %NC', ascending=False).head(10))
print("col7_df : ")
display(col7_df)
```

```python
# From here down is for Activity 3


# 1. Generate a confusion matrix for the two given mutations - RNF43 and RPL22
# Note : Line 230 in the feature table is the RNF43_GRCh37_17:56435161-56435161_Frame-Shift-Del_DEL_C-C-- mutation
#        Line 405 in the feature table is the RPL23AP82_GRCh37_22:51237600-51237600_RNA_DEL_A-A-- mutation
# display(feature_table_df.loc[229:231,:])
# display(feature_table_df.loc[404:406,:])

#Get the data from the table into variables first
TP = feature_table_df.loc[230,'C']
FP = feature_table_df.loc[230,'NC']
FN = total_c - feature_table_df.loc[230,'C']
TN = total_nc - feature_table_df.loc[230,'NC']

#Put the data into a dataframe and print it
conf_matrix_RNF43 = { 'Predicted Positive (PP)' : [TP, FN] , 'Predicted Negative (PN)' : [FP, TN] }
conf_matrix_RNF43 = pd.DataFrame(conf_matrix_RNF43, index={'Actual Positive (P)', 'Actual Negative (N)'})
print("Confusion matrix for RNF43 : ")
print(conf_matrix_RNF43)

#Then do the same for the other mutation
TP = feature_table_df.loc[405,'C']
FP = feature_table_df.loc[405,'NC']
FN = total_c - feature_table_df.loc[405,'C']
TN = total_nc - feature_table_df.loc[405,'NC']

conf_matrix_RPL23 = { 'Predicted Positive (PP)' : [TP, FN] , 'Predicted Negative (PN)' : [FP, TN] }
conf_matrix_RPL23 = pd.DataFrame(conf_matrix_RPL23, index={'Actual Positive (P)', 'Actual Negative (N)'})
print("\nConfusion matrix for RPL23 : ")
print(conf_matrix_RPL23)
                                                                 
```

```python
# 2. Generate stacked bar charts from the confusion matrices

#For RNF43
fig, plot1 = plt.subplots()
plot1.bar(['Positive', 'Negative'], [conf_matrix_RNF43.iloc[0,0], conf_matrix_RNF43.iloc[1,1]], label='True')
plot1.bar(['Positive', 'Negative'], [conf_matrix_RNF43.iloc[0,1], conf_matrix_RNF43.iloc[1,0]], bottom=[conf_matrix_RNF43.iloc[0,0], conf_matrix_RNF43.iloc[1,1]], label='False')
plot1.legend()
plot1.set_title('RNF43 Confusion Matrix')

#For RPL23
fig, plot2 = plt.subplots()
plot2.bar(['Positive', 'Negative'], [conf_matrix_RPL23.iloc[0,0], conf_matrix_RPL23.iloc[1,1]], label='True', color='g')
plot2.bar(['Positive', 'Negative'], [conf_matrix_RPL23.iloc[0,1], conf_matrix_RPL23.iloc[1,0]], bottom=[conf_matrix_RPL23.iloc[0,0], conf_matrix_RPL23.iloc[1,1]], label='False', color='m')
plot2.legend()
plot2.set_title('RPL23 Confusion Matrix')
```

```python
# 3. Generate donut charts from the confusion matrices

#For RNF43
RNF43_list = [conf_matrix_RNF43.iloc[0,0], conf_matrix_RNF43.iloc[0,1], conf_matrix_RNF43.iloc[1,1], conf_matrix_RNF43.iloc[1,0]]
labels = ['TP', 'FP', 'TN', 'FN']
plt.pie(RNF43_list, labels=labels, autopct='%1.1f%%', pctdistance=0.79, colors=['indianred', 'cornflowerblue', 'green', 'gold'])
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('RNF43 Confusion Matrix')
plt.show()

#For RPL23
RPL23_list = [conf_matrix_RPL23.iloc[0,0], conf_matrix_RPL23.iloc[0,1], conf_matrix_RPL23.iloc[1,1], conf_matrix_RPL23.iloc[1,0]]
plt.pie(RPL23_list, labels=labels, autopct='%1.1f%%', pctdistance=0.79, colors=['darkorange', 'cyan', 'lime', 'mediumpurple'])
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('RPL23 Confusion Matrix')
plt.show()
```
