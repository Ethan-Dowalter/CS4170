```python
#Import data file
import pandas as pd
import numpy as np
rawData = pd.read_csv("mutations.csv")
rawData.rename(columns={"Unnamed: 0": "Individual"}, inplace=True)
# display(rawData)
```

```python
# Activity 2


#Calculate totals for each row and column and load it into a dataframe
#Summing up the rows, keeping the individual labels in column 1
num_mut_per_ind = rawData.sum(axis=1, numeric_only=True)
df1 = num_mut_per_ind.to_frame()
df1.insert(0, "Individual", rawData['Individual'])
df1.rename(columns={0 : "Total Mutations"}, inplace=True)
#Summing up the columns, keeping the mutations in column 1
num_ind_per_mut = rawData.sum(numeric_only=True)
df2 = num_ind_per_mut.to_frame()
df2.reset_index(inplace=True)
df2.rename(columns={"index": "Mutation", 0 : "Total Individuals"}, inplace=True)
display(df1)
display(df2)

#Plot number of mutations per individual
import matplotlib.pyplot as plt
plt.scatter(df1.index, df1.loc[:,"Total Mutations"])
plt.title("Number of Mutations Per Individual")
plt.xlabel("Individual")
plt.ylabel("Mutations")
plt.show()

#Plot number of individuals per mutation
plt.scatter(df2.index, df2.loc[:,"Total Individuals"])
plt.title("Number of Individuals Per Mutation")
plt.xlabel("Mutation")
plt.ylabel("Individuals")
plt.show()

#Construct a feature table with additional analysis on the data
feature_table_df = df2.copy()
#First column T = total individuals per mutation is already calculated
feature_table_df.rename(columns={"Mutation" : "Genetic Mutation (Feature)", "Total Individuals" : 'T'}, inplace=True)
# C = number of individuals with cancer with that mutation
feature_table_df['C'] = 0
# NC = number of non-cancer individuals with that mutation
feature_table_df['NC'] = 0
# %C = % of total individuals with that mutation with cancer
feature_table_df['%C'] = 0
# %NC = % of total individuals with that mutation without cancer
feature_table_df['%NC'] = 0
# %C - %NC = Difference between %C and %NC
feature_table_df['%C - %NC'] = 0
# %C / %NC = Ratio of %C to %NC
feature_table_df['%C / %NC'] = 0
#feature_table_df

#Count up how many individuals with cancer have each mutation
#Axis 0 = vertical axis
for x in range(len(rawData.axes[0])):
    #Axis 1 = horizontal axis
    for y in range(len(rawData.axes[1])):
        #Check for mutation and if the individual has cancer
        if rawData.iloc[x,y] == 1 and rawData.iloc[x,0][:1] == 'C':
            feature_table_df.loc[y-1,'C'] += 1
            
#Get total number of cancer and non-cancer individuals
total_c = 0
for x in range(len(rawData.axes[0])):
    #Check if individual is a C
    if rawData.iloc[x,0][:1] == 'C':
        total_c += 1
        
total_nc = len(rawData.axes[0]) - total_c

# print(total_c)
# print(total_nc)

#From that information, the rest of the table can be calculated
for y in range(len(feature_table_df.axes[0])):
    feature_table_df.loc[y,'NC'] = feature_table_df.loc[y,'T'] - feature_table_df.loc[y,'C']
    feature_table_df.loc[y,'%C'] = feature_table_df.loc[y,'C'] / total_c * 100
    feature_table_df.loc[y,'%NC'] = feature_table_df.loc[y,'NC'] / total_nc * 100
    feature_table_df.loc[y,'%C - %NC'] = feature_table_df.loc[y,'%C'] - feature_table_df.loc[y,'%NC']
    if feature_table_df.loc[y,'%NC'] != 0:
        feature_table_df.loc[y,'%C / %NC'] = feature_table_df.loc[y,'%C'] / feature_table_df.loc[y,'%NC']
    else:
        #Use some small number to avoid dividing by 0
        feature_table_df.loc[y,'%C / %NC'] = feature_table_df.loc[y,'%C'] / 0.000001

#Write out the data to a csv
# feature_table_df.to_csv("feature_table.csv", index=True)

#Format the data to display more clearly here
pd.options.display.float_format = '{:.2f}'.format
feature_table_df = feature_table_df.astype({"T":'int', "C":'int', "NC":'int'}) 
print("feature_table_df : ")
display(feature_table_df)

#Construct a summary feature table, showing the top 10 features (mutations) of each column
#First column, sorted by T
col1_df = pd.DataFrame(feature_table_df.sort_values('T', ascending=False).head(10))
print("col1_df : ")
display(col1_df)

#Second column, sorted by C
col2_df = pd.DataFrame(feature_table_df.sort_values('C', ascending=False).head(10))
print("col2_df : ")
display(col2_df)

#Third column, sorted by NC
col3_df = pd.DataFrame(feature_table_df.sort_values('NC', ascending=False).head(10))
print("col3_df : ")
display(col3_df)

#Fourth column, sorted by %C
col4_df = pd.DataFrame(feature_table_df.sort_values('%C', ascending=False).head(10))
print("col4_df : ")
display(col4_df)

#Fifth column, sorted by %NC
col5_df = pd.DataFrame(feature_table_df.sort_values('%NC', ascending=False).head(10))
print("col5_df : ")
display(col5_df)

#Sixth columnm, sorted by %C - %NC
col6_df = pd.DataFrame(feature_table_df.sort_values('%C - %NC', ascending=False).head(10))
print("col6_df : ")
display(col6_df)

#Seventh column, sorted by %C / %NC
col7_df = pd.DataFrame(feature_table_df.sort_values('%C / %NC', ascending=False).head(10))
print("col7_df : ")
display(col7_df)
```

```python
# Acitivty 4 (part 2)


# Some useful functions for lengthening and using the decision tree

# filterData - takes a list of samples from original dataset and returns a dataframe with data from only those samples
def filterData(list, rawData):
    #iterator for counting through the given list of samples
    iterator = 0;
    #set the columns
    df = pd.DataFrame(columns=rawData.columns);
    #go through rawData and only keep the rows with samples that are in the given list
    for y in range(len(rawData.axes[1])):
        if iterator < len(list) and rawData.iloc[y,0] == list[iterator]:
            iterator += 1
            df.loc[len(df.index)] = rawData.iloc[y,:]    
    return df

# condenseData - calculates and returns a new dataframe with all the totals for the samples and mutations for a given sample set
def condenseData(df):
    #calculate total column
    temp = df.sum()
    #convert to dataframe
    newdf = temp.to_frame()
    #remove first row of junk
    newdf = newdf.iloc[1:]
    #reset the index so mutations are not indices
    newdf.reset_index(inplace=True)
    newdf.rename(columns={'index': 'Mutation', 0 : 'T'}, inplace=True)
    #make new columns
    newdf['C'] = 0
    newdf['NC'] = 0
    newdf['TP - FP'] = 0
    #search for positive samples and add those to the newdf
    for x in range(len(df.axes[0])):
        for y in range(len(df.axes[1])):
            if df.iloc[x,y] == 1 and df.iloc[x,0][:1] == 'C':
                newdf.loc[y-1,'C'] += 1
    #fill in the rest of the table given the number of positive samples
    for x in range(len(newdf.axes[0])):
        newdf.loc[x,'NC'] = newdf.loc[x,'T'] - newdf.loc[x,'C']
        newdf.loc[x,'TP - FP'] = newdf.loc[x,'C'] - newdf.loc[x,'NC']
    #fix formatting
    newdf = newdf.astype({'T':'int', 'C':'int', 'NC':'int', 'TP - FP':'int'}) 
    return newdf

# makeConfusionMatrix - takes a list of positive samples and negative samples and returns a confusion matrix based on them
def makeConfusionMatrix(positives, negatives):
    #Get the data from the lists into variables first
    TP = 0
    FP = 0
    FN = 0
    TN = 0

    #Look through positives list for True Positives and False Positives
    for x in range(len(positives)):
        if positives[x][:1] == 'C':
            TP += 1
        else:
            FP += 1

    #Look through negatives list for True Negatives and False Negatives
    for x in range(len(negatives)):
        if negatives[x][:1] == 'N':
            TN += 1
        else:
            FN += 1

    #Put the data into a dataframe and return it
    conf_matrix = pd.DataFrame(index={'Actual Positive (P)', 'Actual Negative (N)'}, 
                               columns={'Predicted Positive (PP)', 'Predicted Negative (PN)'})
    conf_matrix = conf_matrix[['Predicted Positive (PP)', 'Predicted Negative (PN)']]
    conf_matrix = conf_matrix.sort_index(ascending=False)
    conf_matrix.iloc[0,0] = TP
    conf_matrix.iloc[0,1] = FN
    conf_matrix.iloc[1,0] = FP
    conf_matrix.iloc[1,1] = TN
    
    return conf_matrix

# isIn - takes a sample and classifies it based on if it is in the list, returns true or false
def isIn(sample, list):
    for x in range(len(list)):
        if sample == list[x]:
            return True
    return False


# 5. Apply steps 1 and 2 from Activity 4 (part 1) to group A

filteredGroupA = filterData(groupA, rawData)
sortedGroupA = condenseData(filteredGroupA).sort_values('TP - FP', ascending=False).head(11)
#Remove first row to avoid redundancy
sortedGroupA = sortedGroupA.iloc[1:]
print('Sorted Group A : ')
display(sortedGroupA)

#As seen in the table, the UBR5 mutation maximizes 'TP - FP' to a value of 7

#Group A1 has the mutation, Group A2 does NOT have the mutation
groupA1 = []
groupA2 = []

#Note: The UBR5 mutation is in column 314
for y in range(len(filteredGroupA.axes[0])):
    if filteredGroupA.iloc[y,314] == 1:
        groupA1.append(filteredGroupA.iloc[y,0])
    else:
        groupA2.append(filteredGroupA.iloc[y,0])

print('Samples with UBR5 : ', groupA1, '\n')
print('Samples without UBR5 : ', groupA2)

display(makeConfusionMatrix(groupA1, groupA2))

# 6. Apply steps 1 and 2 from Activity 4 (part 1) to group B

filteredGroupB = filterData(groupB, rawData)
sortedGroupB = condenseData(filteredGroupB).sort_values('TP - FP', ascending=False).head(10)
print('Sorted Group B : ')
display(sortedGroupB)

#As seen in the table, the ACVR2A mutation maximizes 'TP - FP' to a value of 5

#Group B1 has the mutation, Group B2 does NOT have the mutation
groupB1 = []
groupB2 = []

#Note: The ACVR2A mutation is in column 173
for y in range(len(filteredGroupB.axes[0])):
    if filteredGroupB.iloc[y,173] == 1:
        groupB1.append(filteredGroupB.iloc[y,0])
    else:
        groupB2.append(filteredGroupB.iloc[y,0])

print('Samples with ACVR2A : ', groupB1, '\n')
print('Samples without ACVR2A : ', groupB2)

display(makeConfusionMatrix(groupB1, groupB2))

# 10. Classify the following samples: C1, C10, C50, NC5, and NC15 using this decision tree

# C1
if isIn('C1', groupA):
    if isIn('C1', groupA1):
        print('C1 is classified as C')
    else:
        print('C1 is classified as NC')
else:
    if isIn('C1', groupB1):
        print('C1 is classified as C')
    else:
        print('C1 is classified as NC')

# C10
if isIn('C10', groupA):
    if isIn('C10', groupA1):
        print('C10 is classified as C')
    else:
        print('C10 is classified as NC')
else:
    if isIn('C10', groupB1):
        print('C10 is classified as C')
    else:
        print('C10 is classified as NC')
        
# C50
if isIn('C50', groupA):
    if isIn('C50', groupA1):
        print('C50 is classified as C')
    else:
        print('C50 is classified as NC')
else:
    if isIn('C50', groupB1):
        print('C50 is classified as C')
    else:
        print('C50 is classified as NC')
        
# NC5
if isIn('NC5', groupA):
    if isIn('NC5', groupA1):
        print('NC5 is classified as C')
    else:
        print('NC5 is classified as NC')
else:
    if isIn('NC5', groupB1):
        print('NC5 is classified as C')
    else:
        print('NC5 is classified as NC')
        
# NC15
if isIn('NC15', groupA):
    if isIn('NC15', groupA1):
        print('NC15 is classified as C')
    else:
        print('NC15 is classified as NC')
else:
    if isIn('NC15', groupB1):
        print('NC15 is classified as C')
    else:
        print('NC15 is classified as NC')
```

```python
# Activity 7 (part 1)


# The Phi Function

# phiFunction - calculates and returns a new dataframe feature table with all the parameters that go into the Phi function
def phiFunction(df):
    #calculate total column
    temp = df.sum()
    #convert to dataframe
    newdf = temp.to_frame()
    #remove first row of junk
    newdf = newdf.iloc[1:]
    #reset the index so mutations are not indices
    newdf.reset_index(inplace=True)
    newdf.rename(columns={'index': 'Mutation', 0 : 'n(t_L)'}, inplace=True)
    #fix formatting (don't need deciaml places for integers)
    newdf = newdf.astype({'n(t_L)':'int'})
    #count how many positive cancer samples are in the dataframe
    positives = 0
    for x in range(len(df)):
        if df.iloc[x,0][:1] == 'C':
            positives += 1
    negatives = len(df) - positives
    #make new columns
    newdf.insert(loc=1, column='n(t)', value=len(df))
    newdf.insert(loc=2, column='n(t, C)', value=positives)
    newdf.insert(loc=3, column='n(t, NC)', value=negatives)
    newdf['n(t_R)'] = newdf['n(t)'] - newdf['n(t_L)']
    newdf['n(t_L, C)'] = 0
    newdf['n(t_L, NC)'] = 0
    newdf['n(t_R, C)'] = 0
    newdf['n(t_R, NC)'] = 0
    #search for positive samples and add those to the newdf
    for x in range(len(df)):
        for y in range(len(df.axes[1])):
            if df.iloc[x,y] == 1 and df.iloc[x,0][:1] == 'C':
                newdf.loc[y-1,'n(t_L, C)'] += 1
    #fill in the rest of the table given the number of positive samples
    newdf['n(t_L, NC)'] = newdf['n(t_L)'] - newdf['n(t_L, C)']
    newdf['n(t_R, C)'] = newdf['n(t, C)'] - newdf['n(t_L, C)']
    newdf['n(t_R, NC)'] = newdf['n(t, NC)'] - newdf['n(t_L, NC)']
    
    #make more columns
    newdf['P_L'] = newdf['n(t_L)'] / newdf['n(t)']
    newdf['P_R'] = newdf['n(t_R)'] / newdf['n(t)']
    newdf['P(C | t_L)'] = newdf['n(t_L, C)'] / newdf['n(t_L)']
    newdf['P(NC | t_L)'] = newdf['n(t_L, NC)'] / newdf['n(t_L)']
    newdf['P(C | t_R)'] = newdf['n(t_R, C)'] / newdf['n(t_R)']
    newdf['P(NC | t_R)'] = newdf['n(t_R, NC)'] / newdf['n(t_R)']
    newdf['Q(s | t)'] = abs(newdf['P(C | t_L)'] - newdf['P(C | t_R)']) + abs(newdf['P(NC | t_L)'] - newdf['P(NC | t_R)'])
    newdf['Phi(s, t)'] = 2 * newdf['P_L'] * newdf['P_R'] * newdf['Q(s | t)']
    
    # Naming convention
    # n(t) - Number of samples in the node
    # n(t, C) - Number of cancer samples in the node
    # n(t, NC) - Number of non cancer samples in the node
    # n(t_L) - Number of samples in the left child node when classified by given mutation
    # n(t_R) - Number of samples in the right child node when classified by given mutation
    # n(t_L, C) - Number of cancer samples in the left child node
    # n(t_L, NC) - Number of NON cancer samples in the left child node
    # n(t_R, C) - Number of cancer samples in the right child node
    # n(t_R, NC) - Number of NON cancer samples in the right child node
    # P_L - probability of sample ending up in left child node
    # P_R - probability of sample ending up in right child node
    # P(C | t_L) - probability of sample having cancer in the left child node
    # P(NC | t_L) - probability of sample NOT having cancer in the left child node
    # P(C | t_R) - probability of sample having cancer in the right child node
    # P(NC | t_R) - probability of sample NOT having cancer in the right child node
    # Q(s | t) - purity measure of a given node
    # Phi(s, t) - resultant value to classify by
    
    return newdf

pd.options.display.float_format = '{:.5f}'.format
display(phiFunction(rawData).sort_values('Phi(s, t)', ascending=False).head(10))
```

```python
# Activity 8 (part 1)


# infoGain - calculates and returns a new dataframe feature table with all the parameters that go into the information gain function
def infoGain(df):
    #calculate total column
    temp = df.sum()
    #convert to dataframe
    newdf = temp.to_frame()
    #remove first row of junk
    newdf = newdf.iloc[1:]
    #reset the index so mutations are not indices
    newdf.reset_index(inplace=True)
    newdf.rename(columns={'index': 'Mutation', 0 : 'n(t_L)'}, inplace=True)
    #fix formatting (don't need deciaml places for integers)
    newdf = newdf.astype({'n(t_L)':'int'})
    #count how many positive cancer samples are in the dataframe
    positives = 0
    for x in range(len(df)):
        if df.iloc[x,0][:1] == 'C':
            positives += 1
    negatives = len(df) - positives
    #make new columns
    newdf.insert(loc=1, column='n(t)', value=len(df))
    newdf.insert(loc=2, column='n(t, C)', value=positives)
    newdf.insert(loc=3, column='n(t, NC)', value=negatives)
    newdf['n(t_R)'] = newdf['n(t)'] - newdf['n(t_L)']
    newdf['n(t_L, C)'] = 0
    newdf['n(t_L, NC)'] = 0
    newdf['n(t_R, C)'] = 0
    newdf['n(t_R, NC)'] = 0
    #search for positive samples and add those to the newdf
    for x in range(len(df)):
        for y in range(len(df.axes[1])):
            if df.iloc[x,y] == 1 and df.iloc[x,0][:1] == 'C':
                newdf.loc[y-1,'n(t_L, C)'] += 1
    #fill in the rest of the table given the number of positive samples
    newdf['n(t_L, NC)'] = newdf['n(t_L)'] - newdf['n(t_L, C)']
    newdf['n(t_R, C)'] = newdf['n(t, C)'] - newdf['n(t_L, C)']
    newdf['n(t_R, NC)'] = newdf['n(t, NC)'] - newdf['n(t_L, NC)']
    newdf['P(C | t)'] = newdf['n(t, C)'] / newdf['n(t)']
    newdf['P(NC | t)'] = newdf['n(t, NC)'] / newdf['n(t)']
    
    #make more columns
    newdf['P_L'] = newdf['n(t_L)'] / newdf['n(t)']
    newdf['P_R'] = newdf['n(t_R)'] / newdf['n(t)']
    newdf['P(C | t_L)'] = newdf['n(t_L, C)'] / newdf['n(t_L)']
    newdf['P(NC | t_L)'] = newdf['n(t_L, NC)'] / newdf['n(t_L)']
    newdf['P(C | t_R)'] = newdf['n(t_R, C)'] / newdf['n(t_R)']
    newdf['P(NC | t_R)'] = newdf['n(t_R, NC)'] / newdf['n(t_R)']
    newdf['H(t)'] = -1 * (newdf['P(C | t)'] * np.log2(newdf['P(C | t)'] + 0.000001) + newdf['P(NC | t)'] * np.log2(newdf['P(NC | t)'] + 0.000001))
    newdf['H(t_L)'] = -1 * (newdf['P(C | t_L)'] * np.log2(newdf['P(C | t_L)'] + 0.000001) + newdf['P(NC | t_L)'] * np.log2(newdf['P(NC | t_L)'] + 0.000001))
    newdf['H(t_R)'] = -1 * (newdf['P(C | t_R)'] * np.log2(newdf['P(C | t_R)'] + 0.000001) + newdf['P(NC | t_R)'] * np.log2(newdf['P(NC | t_R)'] + 0.000001))
    newdf['H(s, t)'] = newdf['P_L'] * newdf['H(t_L)'] + newdf['P_R'] * newdf['H(t_R)']
    newdf['Gain(s)'] = newdf['H(t)'] - newdf['H(s, t)']
    
    # Naming convention
    # n(t) - Number of samples in the node
    # n(t, C) - Number of cancer samples in the node
    # n(t, NC) - Number of non cancer samples in the node
    # P(C | t) - Probability of sample having cancer at node t
    # P(NC | t) - Probability of sample NOT having cancer at node t
    # n(t_L) - Number of samples in the left child node when classified by given mutation
    # n(t_R) - Number of samples in the right child node when classified by given mutation
    # n(t_L, C) - Number of cancer samples in the left child node
    # n(t_L, NC) - Number of NON cancer samples in the left child node
    # n(t_R, C) - Number of cancer samples in the right child node
    # n(t_R, NC) - Number of NON cancer samples in the right child node
    # P_L - Probability of sample ending up in left child node
    # P_R - Probability of sample ending up in right child node
    # P(C | t_L) - Probability of sample having cancer in the left child node
    # P(NC | t_L) - Probability of sample NOT having cancer in the left child node
    # P(C | t_R) - Probability of sample having cancer in the right child node
    # P(NC | t_R) - Probability of sample NOT having cancer in the right child node
    # H(t) - Entropy at node t
    # H(t_L) - Entropy at left child node
    # H(t_R) - Entropy at right child node
    # H(s, t) - Entropy of candidate split s at node t
    # Gain(s) - Information gain of split s at node t
    
    return newdf

pd.options.display.float_format = '{:.5f}'.format
pd.set_option('max_columns', None)
display(infoGain(rawData).sort_values('Gain(s)', ascending=False).head(10))
```

```python
# Activity 10


pd.set_option('max_rows', None)
# Build deeper decision trees and grow the random forest

# makeDeepRandPhiDecisionTree - same thing as makeRandPhiDecisionTree but has makes the tree deeper than depth 2
def makeDeepRandPhiDecisionTree(df, n, seed, depth):
    print(depth, end=" ")
    #Sample only n features, making sure to add back 'Individual' column
    newdf = featureSample(df, n, seed)
    #Root node
    bestFeature = phiFunction(newdf).sort_values('Phi(s, t)', ascending=False).head(5)
    # display(bestFeature)
    bestCol = bestFeature.iloc[0,0]
    root = Node(bestFeature.iloc[0,0])
    group_pos = pd.DataFrame(columns=df.columns)
    group_neg = pd.DataFrame(columns=df.columns)
    for x in range(len(newdf)):
        if newdf.loc[x, bestCol] == 1:
            group_pos = group_pos.append(newdf.iloc[x])
        else:
            group_neg = group_neg.append(newdf.iloc[x])
    #Make sure to drop the feature from the list so that it can be skipped for future nodes
    group_pos.drop(columns=bestCol, inplace=True)
    group_neg.drop(columns=bestCol, inplace=True)
    
    #Left child - positive group
    if len(group_pos) > 0:
        if isPure(group_pos):
            # Base case
            if group_pos.iloc[0,0][:1] == 'C':
                root.left = Node("C")
                # print("pos : C")            
            else:
                root.left = Node("NC")
                # print("pos : NC")            
            # print(root.data)
            # print("Positives : ")
            # for x in range(len(group_pos)):
            #     print(group_pos.iloc[x,0], end=" ")
        else:
            root.left = makeDeepRandPhiDecisionTree(group_pos, n, seed, depth + 1)
    else:
        root.left = Node("C")
    
    #Right child - negative group
    if isPure(group_neg):
        # Base case
        if group_neg.iloc[0,0][:1] == 'N':
            root.right = Node("NC")
            # print("neg : NC")            
        else:
            root.right = Node("C")
            # print("neg : C")
        # print(root.data)
        # print("Negatives : ")
        # for x in range(len(group_neg)):
        #     print(group_neg.iloc[x,0], end=" ")
    else:
        # Check number of cancer samples in right node, base case is when it is <= 5
        # if bestFeature.iloc[0,8] > 5:
        if depth < 5 and bestFeature.iloc[0,8] > 5:
            root.right = makeDeepRandPhiDecisionTree(group_neg, n, seed, depth + 1)
        else:
            root.right = Node("NC")
            # print(root.data)
            # print("Negatives : ")
            # for x in range(len(group_neg)):
            #     print(group_neg.iloc[x,0], end=" ")

    return root

# printTree - prints a binary tree with a pre-order traversal (root -> left -> right)
def printTree(root):
    if root != None:
        print(root.data)
        printTree(root.left)
        printTree(root.right)

# myClassifier - classifies a single sample given a decision tree
def myClassifier(root, sample):
    if root.data == 'C':
        return 'C'
    elif root.data == 'NC':
        return 'NC'
    elif sample[root.data] == 1:
        return myClassifier(root.left, sample)
    else:
        return myClassifier(root.right, sample)

# myRFClassifier - classifies a single sample given a random forest
def myRFClassifier(rootList, sample):
    Cvotes = 0
    NCvotes = 0
    for x in range(len(rootList)):
        if myClassifier(rootList[x], sample) == 'C':
            Cvotes += 1
        else:
            NCvotes += 1
    if Cvotes > NCvotes:
        return 'C'
    else:
        return 'NC'

# makeRF - makes a random forest with 'size' number of trees and 'num_features' number of random features per node
def makeRF(df, size, num_features):
    bootstrapList = []
    rootList = []
    for i in range(size):
        bootstrapList.append(Bootstrap(df, i))
        rootList.append(makeDeepRandPhiDecisionTree(bootstrapList[i][0], num_features, i+45, 1))
        print('\nrootList[', i, ']')
        # printTree(rootList[i])
        
    return bootstrapList, rootList

# cullData - takes the rawData df and removes any features with less than 'cutoff' number of samples
def cullData(df, cutoff):
    culledData = df.copy()
    for i in range(len(culledData.columns) - 1, 0, -1):
        num_samples = 0
        has_Csamples = 0
        for j in range(len(culledData)):
            if culledData.iloc[j,i] == 1:
                num_samples += 1
                if culledData.iloc[j,0][:1] == 'C':
                    has_Csamples = 1
        if num_samples < cutoff:
            culledData.drop(columns=culledData.columns[i], inplace=True)
        elif has_Csamples == 0:
            culledData.drop(columns=culledData.columns[i], inplace=True)
    return culledData

# countMutations - counts the number of times a mutation occurs at the root of a tree
def countMutations(rootList):
    df = pd.DataFrame(columns={'Mutation', 'Count'})
    df = df[['Mutation', 'Count']]
    count = 0
    for i in range(len(rootList)):
        if rootList[i].data not in df.values:
            df.loc[count] = [rootList[i].data, 1]
            count += 1
        else:
            for j in range(len(df)):
                if rootList[i].data == df.iloc[j,0]:
                    df.iloc[j,1] += 1
    return df.sort_values('Count', ascending=False)


temp = cullData(set12, 3)
# display(temp)
bootstrapList, rootList = makeRF(temp, 99, math.floor(len(temp.columns) / 3))

pos = []
neg = []
for i in range(len(set3)):
    if myRFClassifier(rootList, set3.loc[i]) == 'C':
        pos.append(set3.iloc[i,0])
    else:
        neg.append(set3.iloc[i,0])
temp = makeConfusionMatrix(pos, neg)
display(temp)
display(computeMetrics(temp))
display(countMutations(rootList))


temp = cullData(set23, 3)
# display(temp)
bootstrapList, rootList = makeRF(temp, 99, math.floor(len(temp.columns) / 3))

pos = []
neg = []
for i in range(len(set1)):
    if myRFClassifier(rootList, set1.loc[i]) == 'C':
        pos.append(set1.iloc[i,0])
    else:
        neg.append(set1.iloc[i,0])
temp = makeConfusionMatrix(pos, neg)
display(temp)
display(computeMetrics(temp))
display(countMutations(rootList))


temp = cullData(set13, 3)
# display(temp)
bootstrapList, rootList = makeRF(temp, 99, math.floor(len(temp.columns) / 3))

pos = []
neg = []
for i in range(len(set2)):
    if myRFClassifier(rootList, set2.loc[i]) == 'C':
        pos.append(set2.iloc[i,0])
    else:
        neg.append(set2.iloc[i,0])
temp = makeConfusionMatrix(pos, neg)
display(temp)
display(computeMetrics(temp))
display(countMutations(rootList))
```
